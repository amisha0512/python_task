### find a website where you can pass parms ,header and auth

```python
import requests as rq
url = 'https://api.first.org/data/v1/countries'
```

```python
Resp = rq.get(url)
Resp
```

```python
print(Resp.content.decode('utf-8'))
```

### Headres

```python
type(Resp.headers)

lHeader = Resp.headers

print(lHeader)
```

```python
print(f'Content-Type = {lHeader["Content-Type"]}')

print(f'Date = {lHeader["Date"]}')
```

```python
userAgent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'
```

```python
cHeader = rq.get(url, headers= {'User-Agent': userAgent })
print(cHeader)
```

```python
cHeader.request.headers
```

### Params

```python
url = 'https://api.first.org/data/v1/countries'
```

```python
cParams = {'code':'CO','name': 'Colombia', 'locations': 60}
Resp = rq.get(url, params=cParams)
print(Resp.json())
```

### Authentication


```python
url = 'https://api.openaq.org/v2/countries'
```

```python
aAuth = ("name","Afghanistan")

type(aAuth)
```

```python
authResp = rq.get(url, auth=aAuth)
authResp
```

```python
authResp.json()
```
